```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px ### for plotting the data on world map


path = 'Death.csv'
df = pd.read_csv(path)
df.info()
df.head(100)
```

```python
df.drop(['Series Name',	'Series Code', 'Country Code',	'1990 [YR1990]', '2001 [YR2001]',	'2012 [YR2012]',	'2013 [YR2013]',	'2014 [YR2014]',	'2016 [YR2016]',	'2017 [YR2017]',	'2018 [YR2018]',	'2020 [YR2020]',	'2021 [YR2021]'], axis=1, inplace = True)
```

```python
df.rename(columns={'Country Name': "Country"}, inplace=True)
df.rename(columns={'2000 [YR2000]': "2000"}, inplace=True)
df.rename(columns={'2015 [YR2015]': "2015"}, inplace=True)
df.rename(columns={'2019 [YR2019]': "2019"}, inplace=True)
df.head(10)
```

```python
world = df.groupby("Country")['2000','2015','2019'].sum().reset_index()
world.head(10)
```

```python
figure = px.choropleth(world,locations= 'Country', locationmode='country names', color='2000', hover_name='Country', color_continuous_scale='tealgrn', range_color=[1,1000000],title='Confirmed cases for individual countries')
figure.show()
```

```python
figure = px.choropleth(world,locations= 'Country', locationmode='country names', color='2015', hover_name='Country', color_continuous_scale='tealgrn', range_color=[1,1000000],title='Confirmed cases for individual countries')
figure.show()
```

```python
figure = px.choropleth(world,locations= 'Country', locationmode='country names', color='2019', hover_name='Country', color_continuous_scale='tealgrn', range_color=[1,1000000],title='Confirmed cases for individual countries')
figure.show()
```
